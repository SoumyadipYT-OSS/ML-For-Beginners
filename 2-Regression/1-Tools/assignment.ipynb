```python
# import Scikit-learn built-in methods and other libraries
from sklearn.datasets import load_linnerud
import numpy as np
```

```python
# load the dataset
data = load_linnerud()
```

```python
X = data.data     # Physical exercise measurements
y = data.target   # Physiological measurements
```

```python
# Focus on Relevant Data
# Select the first two features and the first two targets
# Input(X): Situps (inex 1 of X)
# Output(y): Waistline (index 1 of y)

situps = X[:, 1].reshape(-1, 1)  # Reshape for compaibility
waistline = y[:, 1]  # Waistline measurements
```

```python
# Train-Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(situps, waistline, test_size=0.2, random_state=42)
```

```python
# Create a Regression Model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
# Fit the model to the training data
model.fit(X_train, y_train)
```

```python
# Visualize the Results
import matplotlib.pyplot as plt
# Plot the training data
plt.scatter(X_train, y_train, color='blue', label='Training Data')
# Plot the test data
plt.scatter(X_test, y_test, color='green', label='Test Data')
# Plot the regression line
plt.plot(X_train, model.predict(X_train), color='red', label='Regression Line')
plt.xlabel('Situps')
plt.ylabel('Waistline')
plt.title('Situps vs Waistline')
plt.legend()
plt.show()
```

```python
# Finally we check the model performance and efficiency
from sklearn.metrics import mean_squared_error, r2_score
# Make predictions on the test set
y_pred = model.predict(X_test)
# Calculate Mean Squared Error
mse = mean_squared_error(y_test, y_pred)
# Calculate R-squared
r2 = r2_score(y_test, y_pred)
print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
```
