```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model, model_selection
```

```python
# Load the diabetes dataset
X, y = datasets.load_diabetes(return_X_y=True)
print(X.shape)
print(X[0])
```

##### Reshape the data to be a 2D array - as required for plotting

```python
X = X[:, 2].reshape((-1, 1))
```

```python
print(X.shape)
```

```python
print(X)
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.33)
```

```python
# Create linear regression object
model = linear_model.LinearRegression()
# Train the model using the training sets
model.fit(X_train, y_train)
```

```python
# Make predictions using the testing set
y_pred = model.predict(X_test)
```

```python
# print y_pred to check the predictions
print(y_pred)
```

```python
# Plot outputs
plt.scatter(X_test, y_test, color="black")
plt.plot(X_test, y_pred, color="blue", linewidth=3)
plt.xlabel("Scaled BMI")
plt.ylabel("Diabetes Progression")
plt.title("Diabetes Progression Prediction")
plt.show()
```

```python
# Advanced Plot Output
plt.figure(figsize=(10, 6))
plt.scatter(X_test, y_test, color="black", label="Test Data")
plt.plot(X_test, y_pred, color="blue", linewidth=3, label="Predicted Line")
plt.xlabel("Scaled BMI")
plt.ylabel("Diabetes Progression")
plt.title("Diabetes Progression Prediction")
plt.legend()
plt.grid()
plt.show()
```

```python
# Advanced Plot Output with Annotations
plt.figure(figsize=(10, 6))
plt.scatter(X_test, y_test, color="black", label="Test Data")
plt.plot(X_test, y_pred, color="blue", linewidth=3, label="Predicted Line")
plt.xlabel("Scaled BMI")
plt.ylabel("Diabetes Progression")
plt.title("Diabetes Progression Prediction")
plt.legend()
plt.grid()
for i, txt in enumerate(y_test):
    plt.annotate(txt, (X_test[i], y_pred[i]), fontsize=8)
plt.show()
```
